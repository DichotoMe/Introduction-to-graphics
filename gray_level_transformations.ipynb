```python
from PIL import Image
from pylab import *
```

```python
def negative(img_arr):
    return 255 - img_arr
```

```python
img = array(Image.open('images/profile.jpg'))
Image.fromarray(negative(img))
```

```python
def clamp(img, start_end):
    start, end = start_end
    diff = end - start
    
    return np.asarray(diff / 255 * img + start).astype(np.uint8)
```

```python
Image.fromarray(clamp(img, (100, 200)))
```
