```python
from PIL import Image
from pylab import *
import numpy as np
```

```python
arr1 = array(Image.open('images/girlface.bmp'))
arr2 = array(Image.open('images/houses.bmp'))
```

```python
def addImages(img1, img2):
    sm = img1.copy() + img2.copy()
    res = (sm - sm.min()) / (sm.max() - sm.min()) * 255
    return np.asarray(res).astype(np.uint8)
```

```python
Image.fromarray(addImages(arr1, arr2))
```
