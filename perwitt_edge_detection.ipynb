```python
import cv2
from skimage import img_as_int, img_as_ubyte
from PIL import Image
import scipy.ndimage.filters as flt
import numpy as np
```

```python
img_gray = img_as_int(cv2.imread('images/moon.jpg', 0))
```

```python
prewitt_vertical = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
Image.fromarray(flt.convolve(img_gray, prewitt_vertical))
```

```python
prewitt_horizontal = prewitt_vertical.swapaxes(0, 1)
```

```python
Image.fromarray(flt.convolve(img_gray, prewitt_horizontal))
```
