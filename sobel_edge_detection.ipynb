```python
import cv2
from skimage import img_as_int, img_as_ubyte
from PIL import Image
import scipy.ndimage.filters as flt
import numpy as np
```

```python
img_gray = img_as_int(cv2.imread('images/profile.jpg', 0))
```

```python
sobel_vertical = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])
```

```python
Image.fromarray(flt.convolve(img_gray, sobel_vertical))
```

```python
sobel_horizontal = sobel_vertical.T
```

```python
Image.fromarray(flt.convolve(img_gray, sobel_horizontal))
```

```python
img_gray = cv2.imread('images/profile.jpg', 0)

sobel_x = cv2.Sobel(img_gray, -1, 1, 0, ksize=5)
sobel_y = cv2.Sobel(img_gray, -1, 0, 1, ksize=5)
```

```python
Image.fromarray(sobel_x)
```
