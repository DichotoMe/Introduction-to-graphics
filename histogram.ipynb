```python
from PIL import Image
from pylab import *
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = Image.open("images/moon_sobel.jpg", ).convert("L")
img
```

```python
# Drawing a histogram

img_array = array(img)

figure()
flattened = img_array.flatten()
hist(flattened, 300, cumulative=True)
title("Histogram")
show()
```

```python
import numpy as np 
import scipy.misc, math
```

```python
img1 = np.asarray(img)
fl = img1.flatten()

histogram, bins = np.histogram(img1, 256, [0, 255])
plot(histogram.cumsum())
title("Cumulative histogram")

cdf = histogram.cumsum()

cdf_m = np.ma.masked_equal(cdf, 0)
```

```python
nom_cdf_m = (cdf_m - cdf_m.min()) * 255
den_cdf_m = (cdf_m.max() - cdf_m.min())

cdf_m = nom_cdf_m / den_cdf_m

cdf = np.ma.filled(cdf_m, 0).astype('uint8')

img2 = cdf[fl]
img3 = np.reshape(img2, img1.shape)

histogram, bins = np.histogram(img3, 256, [0, 255])
plot(histogram.cumsum())
title("Cumulative histogram")

img4 = Image.fromarray(img3)
img4
```

```python
figure()
arr = array(img4)
hist(arr.flatten(), 300)
title("Histogram")
show()
```
