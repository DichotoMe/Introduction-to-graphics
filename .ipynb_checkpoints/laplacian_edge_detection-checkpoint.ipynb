```python
import cv2
from skimage import img_as_int, img_as_ubyte
from PIL import Image
import scipy.ndimage.filters as flt
import numpy as np
```

```python
img_gray = cv2.imread('images/profile.jpg', 0)
```

```python
laplacian_negative = np.array([
    [0, -1, 0], 
    [-1, 4, -1], 
    [0, -1, 0]
])
```

```python
Image.fromarray(flt.convolve(img_as_int(img_gray), laplacian_negative))
```

```python
laplacian_positive = np.array([
    [0, 1, 0], 
    [1, -4, 1], 
    [0, 1, 0]
])
```

```python
Image.fromarray(flt.convolve(img_as_int(img_gray), laplacian_positive))
```

```python
laplacian = cv2.Laplacian(img_gray, -1)

Image.fromarray(laplacian)
```
